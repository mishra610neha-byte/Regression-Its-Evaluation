**Question 1:  What is Simple Linear Regression?**

**Answer**: Simple Linear Regression means finding a straight line that best shows the relationship between one input (X) and one output (Y).       

 **example**:             
If you look at hours studied (X) and marks scored (Y), the model finds the best straight line that shows:

**‚ÄúMore** **hours** **‚Üí** more marks.

**Question 2: What are the key assumptions of Simple Linear Regression?**

**Answer**:
1. **Linearity**

The model assumes a direct, strategic alignment between the independent variable (X) and the dependent variable (Y). In short, the relationship is expected to follow a straight-line trajectory.

2. **Independence** **of** **Errors**

Residuals should function as standalone operational units‚Äîno cross-talk, no dependencies. Each error term must be independent to maintain analytical integrity.

3. **Homoscedasticity**

Error variance is expected to remain consistent across the entire X spectrum. In other words, the model anticipates a stable noise floor without volatility spikes.

4. **Normality** **of** **Residuals**

Residuals should follow a normal distribution curve. This enables the downstream statistical insights‚Äîlike confidence intervals and hypothesis tests‚Äîto remain investment-grade.

5. **No** **Significant** **Outliers**

Outliers can create strategic distortions in the regression line. The framework assumes the data environment is relatively clean and outlier-light.

6. **X is Measured Without Error**

The independent variable is presumed to come from a reliable, well-governed data source with minimal measurement friction.

**Question 6:  Implement a Python program to fit a Simple Linear Regression model** **to**       
The following sample data:     
‚óè **X** = [**1**, **2**, **3**, **4**, **5**]      
**‚óè Y = [2.1, 4.3, 6.1, 7.9, 10.2]**     
**Plot the regression line over the data points.**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Core data assets
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

# Model instantiation and deployment
model = LinearRegression()
model.fit(X, Y)

# Strategic forecasts
Y_pred = model.predict(X)

print("Slope (Coefficient):", model.coef_[0])
print("Intercept:", model.intercept_)

# Visualization layer
plt.scatter(X, Y, label="Actual Data Points")
plt.plot(X, Y_pred, label="Regression Line", linewidth=2)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Simple Linear Regression ‚Äì Performance Visualization")
plt.legend()
plt.tight_layout()
plt.show()
```

**Question** **7**: **Fit a Multiple Linear Regression model on this sample** **data**:          
‚óè Area = [1200, 1500, 1800, 2000]           
‚óè Rooms = [2, 3, 3, 4]        
‚óè Price = [250000, 300000, 320000, 370000]

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

# ---------------------------------------------
# Data Ingestion Layer
# ---------------------------------------------
area = [1200, 1500, 1800, 2000]
rooms = [2, 3, 3, 4]
price = [250000, 300000, 320000, 370000]

df = pd.DataFrame({
    "Area": area,
    "Rooms": rooms,
    "Price": price
})

# ---------------------------------------------
# Model Enablement: Multiple Linear Regression
# ---------------------------------------------
X = df[["Area", "Rooms"]]
X = sm.add_constant(X)   # Adding intercept for model calibration
y = df["Price"]

model = sm.OLS(y, X).fit()
print("=== Regression Summary ===")
print(model.summary())

# ---------------------------------------------
# Multicollinearity Audit: VIF Assessment
# ---------------------------------------------
vif_df = pd.DataFrame()
vif_df["Feature"] = X.columns
vif_df["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

print("\n=== Variance Inflation Factor (VIF) Report ===")
print(vif_df)

```

**Question 8:  Implement polynomial regression on the following** **data**:                
‚óè X = [1, 2, 3, 4, 5]               
‚óè Y = [2.2, 4.8, 7.5, 11.2, 14.7]             
**Fit a 2nd-degree polynomial and plot the resulting curve**.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Input datasets
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

# Generating 2nd-degree polynomial features
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Model deployment
model = LinearRegression()
model.fit(X_poly, Y)

# Predictions for plotting
X_range = np.linspace(1, 5, 100).reshape(-1, 1)
X_range_poly = poly.transform(X_range)
Y_pred = model.predict(X_range_poly)

# Visualization deliverable
plt.scatter(X, Y)
plt.plot(X_range, Y_pred)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("2nd Degree Polynomial Regression Curve")
plt.show()

# Display model coefficients for transparency
print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)
```

**Question** 9: **Create a residuals plot for a regression model trained on this** **data**:            
‚óè X = [10, 20, 30, 40, 50]                 
‚óè Y = [15, 35, 40, 50, 65]                          
**Assess heteroscedasticity by examining the spread of residuals** .

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Input data
X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
Y = np.array([15, 35, 40, 50, 65])

# Model deployment
model = LinearRegression()
model.fit(X, Y)

# Predictions & residual analytics
y_pred = model.predict(X)
residuals = Y - y_pred

# Residuals plot ‚Äî a visual QA checkpoint
plt.scatter(X, residuals)
plt.axhline(y=0)
plt.xlabel("X Values")
plt.ylabel("Residuals")
plt.title("Residuals Plot for Heteroscedasticity Assessment")
plt.show()

print("Residuals:", residuals)
```
